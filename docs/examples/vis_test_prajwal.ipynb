```python
columns = ['id', 'matching_id', 'process', 'receive_from', 'send_to', 'time']
new_df = pd.DataFrame(columns=columns)

for index, row in mpi_trace.iterrows():
    if (row['Name'] == 'MpiSend'):
        print(f"Index: {index}, Name: {row['Name']}, Process: {row['Process']}, Matching: {row['_matching_event']}")
        process = row['Process']
        matching_process = mpi_trace.loc[row['_matching_event']]['Process']
        send_time = row['Timestamp (ns)']
        recv_time = row['_matching_timestamp']
        
        new_df = pd.concat([new_df, pd.DataFrame([{'id': index, 'matching_id': row['_matching_event'], 'process': process, 'receive_from': None, 'send_to': matching_process, 'time': send_time }])], ignore_index=True)
        new_df = pd.concat([new_df, pd.DataFrame([{'id': row['_matching_event'], 'matching_id': index, 'process': matching_process, 'receive_from': process, 'send_to': None, 'time': recv_time }])], ignore_index=True)

sorted_df = new_df.sort_values(by='time')        
print(sorted_df)
```

```python
unique_values_process = sorted_df['process'].unique()
print(unique_values_process)
```

```python
from pipit.partition.partitions import Event

start_event_list = []
event_dict = {}
for process in unique_values_process:
    filtered_df = sorted_df[(sorted_df['process'] == process)]
    # Create start event - We should ideally get start time of each process. For now, I am assuming that all processes start at time 0
    prev_event = Event('Start',process, 0, process, 0)
    start_event_list.append(prev_event)
    
    #event_type, event_id, time, process, matching_id
    for index, row in filtered_df.iterrows():
        event = None
        if row['send_to'] is not None:
            event = Event('MpiSend', row['id'], row['time'], process, row['matching_id'])
            event.add_prev(prev_event)
            prev_event.add_next(event)
            event_dict[row['id']] = event

            
            prev_event = event
        elif row['receive_from'] is not None:
            event = Event('MpiRecv', row['id'], row['time'], process, row['matching_id'])
            event.add_prev(prev_event)
            prev_event.add_next(event)
            event_dict[row['id']] = event
            
            prev_event = event

    
for se in start_event_list:
    temp = se
    while temp is not None:
        if temp.matching_event_id != 0:
            temp.add_matching(event_dict[temp.matching_event_id])
        temp = temp.next_event
    

for se in start_event_list:
    temp = se
    while temp is not None:
        if temp.matching_event is not None:
            print (f"({temp.process}, {temp.event_id}, {temp.event_type}, {temp.matching_event.process})", end = '->')
        else:
            print (f"({temp.process}, {temp.event_id}, {temp.event_type})", end = '->')
        temp = temp.next_event
    print()
```

```python
from graphviz import Digraph

# Function to create a DOT representation of linked lists
def linked_lists_to_dot(lists):
    dot = Digraph(comment='Linked Lists')

    for i, ll in enumerate(lists):
        current_node = ll
        while current_node:
            dot.node(f"{current_node.event_id}_{current_node.event_type}", str(current_node.event_type) + ":" + str(current_node.event_id))
            if current_node.next_event:
                dot.edge(f"{current_node.event_id}_{current_node.event_type}", f"{current_node.next_event.event_id}_{current_node.next_event.event_type}")
            if current_node.event_type == 'MpiSend' and current_node.matching_event is not None:
                dot.edge(f"{current_node.event_id}_{current_node.event_type}", f"{current_node.matching_event.event_id}_{current_node.matching_event.event_type}")
            current_node = current_node.next_event

    return dot

 # Sample linked lists

# Create a DOT representation
dot_representation = linked_lists_to_dot(start_event_list)
dot_representation.attr(rankdir='LR')  
# Display the DOT representation in a Jupyter Notebook (optional)
dot_representation
```

```python
# Implement Merging of Partitions
from pipit.partition.partitions import Partition

partition_list = []
event_to_partition = {} # event_id -> partition_id
start_partitions = []

# Create a singleton partition for every event
for start_event in start_event_list:
    event = start_event
    while event is not None:
        if event.matching_event is not None:
            if event.matching_event_id in event_to_partition.keys():
                partition_list[event_to_partition[event.matching_event_id]].add_event(event)
                event_to_partition[event.event_id] = event_to_partition[event.matching_event_id]
            else:
                partition_id = len(partition_list)
                partition_list.append(Partition(partition_id, [event]))
                event_to_partition[event.event_id] = partition_id
        else:
            partition_id = len(partition_list)
            partition_list.append(Partition(partition_id, [event]))
            event_to_partition[event.event_id] = partition_id
        if event.event_type == "Start":
            start_partitions.append(partition_list[event_to_partition[event.event_id]])
        #event.add_partition(partition_list[event_to_partition[event.event_id]])
        event = event.next_event

print (len(partition_list))

for partition in partition_list:
    print (f"{partition.partition_id} - {partition.event_list}")
```

```python
from graphviz import Digraph

visited = []

# Function to create a DOT representation of linked lists
def visualise(start_nodes, dot=None):
    if dot is None:
        dot = Digraph(comment='Partition Graph')

    for i, ll in enumerate(start_nodes):
        current_partition = ll
        if  current_partition.partition_id in visited:
            return dot
        dot.node(f"{current_partition.partition_id}", str(current_partition.partition_id))
        visited.append(current_partition.partition_id)
        children = current_partition.get_children()
        for child in children:
            child_partition = partition_list[child]
            dot = visualise([child_partition], dot)
            dot.edge(f"{current_partition.partition_id}", f"{child_partition.partition_id}")
    return dot

 # Sample linked lists

# Create a DOT representation
dot_representation = visualise(start_partitions)
dot_representation.attr(rankdir='LR')  
# Display the DOT representation in a Jupyter Notebook (optional)
dot_representation
```
